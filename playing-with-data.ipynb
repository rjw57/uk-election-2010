```python
import json
```

```python
westminster_constituencies = json.load(open('gb_wpc_2010_05.json'))
```

```python
all_constituencies = {
    'type': 'FeatureCollection',
    'features': westminster_constituencies['features'] #+ scotland_and_wales_constituencies['features'] + greater_london_constituencies['features']
}
all_features = all_constituencies['features']
```

```python
import re
```

```python
all_names = list(c['properties']['name'] for c in all_features)
print('Found %s constituencies in the shape files' % (len(all_names),))

all_by_name = {}
for c in all_features:
    n = c['properties']['name']
    n = n.replace(' Boro', '')
    n = n.replace(' Const', '')
    n = n.replace(' Co', '')
    n = n.strip()
    
    if n.startswith('North '):
        n = n.replace('North ','') + ' North'
    if n.startswith('South '):
        n = n.replace('South ','') + ' South'
    if n.startswith('East '):
        n = n.replace('East ','') + ' East'
    if n.startswith('West '):
        n = n.replace('West ','') + ' West'
    if n.startswith('City of '):
        n = n.replace('City of ','') + ', City of'
    all_by_name[n] = c
    
all_names = all_by_name.keys()
```

```python
sorted(all_by_name.keys())
```

```python
import csv
import difflib
```

```python
r = csv.reader(open('uk-ge-2010-results.csv'))
results_header = r.next()
results = list(r)
print(', '.join(results_header))
```

```python
results_const_names = list(r[1] for r in results)
results_const_regions = list(r[2] for r in results)
print('Found %s constituencies in the results.' % (len(results_const_names),))
```

```python
party_codes = results_header[6:]
```

```python
def votes(row):
    out = {}
    for code, c in zip(party_codes, row[6:]):
        c = c.strip()
        if c == '':
            continue
        out[code] = int(c)
    return out
```

```python
from fuzzywuzzy import process

def match_name(name, region, match_names):
    if region == 'Northern Ireland':
        return None
    dm = difflib.get_close_matches(name, match_names, 1)
    if len(dm) > 0:
        return dm[0], all_by_name[dm[0]]
    fm = process.extractOne(name, match_names)
    if fm is not None:
        return fm[0], all_by_name[fm[0]]
    
    print('Failed with ' + name)
    return None
```

```python
results_matches = []
match_names = list(x for x  in all_names)
for x in results:
    name, region, v, f = (x[1], x[2], votes(x), match_name(x[1], x[2], match_names))
    if f is not None:
        match_names.remove(f[0])
    results_matches.append((name, region, v, f[1] if f is not None else None))
```

```python
print(match_names)
```

```python
print('\n'.join("%s, %s -> %s (%s)" % (n, r, f['properties']['name'] if f is not None else None, ', '.join('%s: %s' % (k,v) for k,v in v.iteritems())) for n, r, v, f in results_matches))
```

```python
matched_constituencies = { 'type': 'FeatureCollection', 'features': [] }
for n, r, v, f in results_matches:
    if f is None:
        continue
    f['properties']['results'] = v
    matched_constituencies['features'].append(f)
```
